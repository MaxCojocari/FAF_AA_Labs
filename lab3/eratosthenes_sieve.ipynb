```python
import matplotlib.pyplot as plt
import numpy as np
import time as t
import math
from prettytable import PrettyTable
```

```python
sample = [True] * 10000
sample_sizes = [1] + [x for x in range(500, 10001, 500)]
```

```python
# Algorithm 1
def alg1(arr: list) -> list:
  c = list(arr)
  c[0] = False
  i = 2
  n = len(c) - 1
  while i <= n:
    if c[i]:
      j = 2 * i
      while j <= n:
        c[j] = False
        j += i
    i += 1
  return c
```

```python
# Algorithm 2
def alg2(arr: list) -> list:
  c = list(arr)
  c[0] = False
  i = 2
  n = len(c) - 1
  while i <= n:
    j = 2 * i
    while j <= n:
      c[j] = False
      j += i
    i += 1
  return c
```

```python
# Algorithm 3
def alg3(arr: list) -> list:
  c = list(arr)
  c[0] = False
  i = 2
  n = len(c) - 1
  while i <= n:
    if c[i]:
      j = i + 1
      while j <= n:
        if j % i == 0:
          c[j] = False
        j += 1
    i += 1
  return c
```

```python
# Algorithm 4
def alg4(arr: list) -> list:
  c = list(arr)
  c[0] = False
  i = 2
  n = len(c) - 1
  while i <= n:
    j = 2
    while j < i:
      if i % j == 0:
        c[i] = False
      j += 1
    i += 1
  return c
```

```python
# Algorithm 5
def alg5(arr: list) -> list:
  c = list(arr)
  c[0] = False
  i = 2
  n = len(c) - 1
  while i <= n:
    j = 2
    while j <= math.sqrt(i):
      if i % j == 0:
        c[i] = False
      j += 1
    i += 1
  return c
```

```python
# testing for correctness
ans = alg1(sample[:100])
for i in range(len(ans)):
  if ans[i]:
    print(i)
```

```python
ellapsed_times = []
```

```python
time_records = []
for n in sample_sizes:
  start = t.time()
  alg1(sample[:n])
  end = t.time()
  time_records.append(end-start)
ellapsed_times.append(time_records)
```

```python
# plot
fig, ax = plt.subplots()

ax.plot(sample_sizes, ellapsed_times[0], linewidth=1.5, color="blue", marker="o", markersize=4.2)
ax.set_xlabel("Array size, n")
ax.set_ylabel("Time, s")
ax.set_title("Algorithm 1")

plt.show()
```
